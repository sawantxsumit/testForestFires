```python
#make a plot from dictionary
nut_prices= { "Almond butter " : 800,
              "Peanut Butter" : 300,
               "cashew butter" : 1000
            }
fig , ax= plt.subplots(2,1, figsize=(5,10)) #5-> width and 10-> height
ax[0].bar(nut_prices.keys(),  nut_prices.values()  );
ax[1].barh(nut_prices.keys(),  nut_prices.values()  ); #horizontal bar graph
ax[0].set(title= "Price of Nut butter" , ylabel="Price(Inr)");
ax[1].set(title= "Price of Nut butter" , xlabel="Price(Inr)");
```

HISTOGRAM

```python
%%time
x= np.random.randn(1000)
print(np.mean(x))
print(np.std(x)) #standard deviation
```

Bins are the number of lines/towers/columns in the graph

Edgecolor is the outline color of the columns

Alpha alters the transparency of the graph

Linestyle is the  style of the cells drawn in the background for better understanding

Grid is the cell in the background like in a graph paper

```python
fig , ax= plt.subplots()
ax.hist(x, bins=30, edgecolor='black', alpha=0.8); 
plt.grid(True, linestyle="-", alpha=0.5);
# ax.hist(x);
```

IMPORTANT ATTRIBUTES OF FIGURE

dpi (dots per inch ) controls the resolution of the graph the more the dpi the sharper and clear the plot but it will increase the filesize

dpi=300 is ideal for printing

fig.patch -> allows customizing the figure background

tight_layout() ->	Adjusts spacing between subplots to prevent overlap

add_subplot()	Adds an extra subplot dynamically (fig.add_subplot(2, 2, 3))

savefig()	Saves the figure as an image file (fig.savefig("plot.png", dpi=300))

IMPORTANT ATTRIBUTES OF AXIS

xlim and ylim are functions used to set the limits (range) of the x-axis and y-axis in a plot. This helps you control what part of the data is shown on each axis.

xlim()  Sets the limits for the x-axis.
ylim()  Sets the limits for the y-axis.

legend()	Displays a legend (ax.legend(["Line 1"]))

set_facecolor()	Changes background color of the subplot

np.random.seed(4) it sets the state of the random number generator so that
any subsequent random numbers generated will be the same every time you run the code.

 The seed does not affect randomness but ensures repeatability.
Any integer can be used as a seed.
Using the same seed will always generate the same numbers.

SUBPLOT

```python
x=set()
np.random.seed(123)
x=np.random.randint(1,10,5)
print(x)
fig ,ax =plt.subplots( 2, 2 , dpi=100)
fig.patch.set(edgecolor="red" , linewidth=3 , facecolor="lightgray")  # Set border color
# fig.patch.set_linewidth(3)  # Set border thickness
# fig.patch.set_facecolor("lightgray")
fig.tight_layout() # can be used when we dont set the figsize


ax[0,0].plot(x);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(x,[2,4,6,8,10] ,edgecolor="Black");
ax[1,1].hist(np.random.randn(10), edgecolor="Black", bins=10);
ax[0,0].set_facecolor("yellow")
ax[0,1].set_facecolor("yellow")
ax[1,0].set_facecolor("yellow")
ax[1,1].set_facecolor("yellow")
ax[0,0].grid(True, linestyle="--", alpha=0.5)
# ax[0,0].legend(["Line 1"]);
fig.savefig("images/4-graph.png")
```

plotting from pandas dataframe

```python
car= pd.read_csv("car-sales.csv")
car
```

```python
np.random.seed(44)
n=np.random.randn(1000)
date= pd.date_range('1/1/2000' , periods= 1000)
ts= pd.Series(n ,index=date )
ts=ts.cumsum()
ts.plot();
```

```python
car["Price"]= car["Price"].str.replace(r'[\$\,\.]', '', regex=True)
#regex means regular expression represented bu r
```

```python
car["Price"]= car["Price"].astype(int)
```

```python
car["Price"]= car["Price"]/100
```

```python
car
```

```python
car["sale date"]= pd.date_range('1/1/2020', periods=10, freq='10D')
car
```

```python
car["total sales"]= car["Price"].cumsum()
```

```python
car.to_csv('car-sales2.csv')
```

```python
car.plot(x="sale date", y= "total sales", kind='area',color='skyblue', linestyle='--', title="Sales over time");
```
