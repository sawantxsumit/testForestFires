#matplotlib OO method

```python
over_50=hd[hd["age"] >50]
len(over_50)
```

```python
over_50.plot.scatter(x='age', y='chol', c='target');
```

```python
#Oject oriented method
fig, ax= plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', 
            x='age',
            y='chol',
            c='target',
            ax=ax);

#xlim and ylim is basically used to zoomin and zoom out the graph we have
#we can view a particular section through this method
ax.set(xlim=[50,60],ylim=[200,300]);
#people of age bw 50-60 who have a cholestrol in range 200-300
```

```python
fig, ax= plt.subplots(figsize=(10,6))
sc=ax.scatter(x=over_50['age'],
          y=over_50['chol'],
          c=over_50['target']);
ax.set(title='Heart disease and cholestrol level',
      xlabel='age',
      ylabel='cholestrol level')
ax.legend(*sc.legend_elements(), title='target');

#add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
over_50
```

```python
# subplot of chol , age,  thalach
fig , ax= plt.subplots(2,1, figsize=(8,5), sharex=True)
fig.patch.set(edgecolor="black", linewidth=3, linestyle='--',facecolor='skyblue')
fig.suptitle("Heart disease", fontsize=15, fontweight='bold')
#add data t0 ax[0]
sc= ax[0].scatter(x=over_50['age'],
                  y= over_50['chol'],
                 c=over_50['target'])
ax[0].set(title='Heart disease and Cholestrol levels',
         ylabel='cholestrol')
ax[0].legend(*sc.legend_elements(), title='target');
#add a meanline
ax[0].axhline(y=over_50['chol'].mean(), linestyle='--')
ax[0].grid(True , linestyle='--', alpha=0.7)

#setting ax[1] axis
sc1=ax[1].scatter(x=over_50['age'],
                 y=over_50['thalach'],
                 c=over_50['target'])
ax[1].set(title='Heart disease and Max heart rate',
       xlabel='Age',
       ylabel='Heart rate')
ax[1].legend(*sc1.legend_elements(), title='target');
ax[1].axhline(over_50['thalach'].mean(), linestyle='--');
ax[1].grid(True , linestyle='--', alpha=0.7)
```
