```python
final_df.groupby(['Aggregate rating' , 'Rating color' , 'Rating text']).size().reset_index()
```

```python
rating=final_df.groupby(['Aggregate rating' , 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
# .rename(columns={0: 'Rating count'}) to rename the column named 0 to rating count
```

```python
rating
```

```python
sns.catplot(x='Aggregate rating', y='Rating count',data=rating[rating['Aggregate rating'] >0], kind='point')
```

#observation
1. when rating is between 4.5 to 4.9 --> Excellent
2. when rating is betwenn 4.0 to 4.5 --> very good
3. when rating is between 3.5 to 4.0 --> good
4. when rating is between 3.0 to 3.5 --> good
5. when rating is between 2.5 to 2.9 --> Average
6. when rating is between 0 to 2.5 --> Poor

```python
import matplotlib
#methods tp change the figsize in seaborn
# matplotlib.rcParams['figure.figsize']=(15,8)
# plt.figure(figsize=(15,8))
sns.set(rc={"figure.figsize": (15, 8)})
sns.set_theme(style="white")
# sns.set_context("paper")  # Options: paper, notebook, talk, poster

sns.barplot(x="Aggregate rating" , y='Rating count', data= rating, hue='Aggregate rating', palette='winter');
```

```python
# Drop NaNs from the 'Cuisines' column
cuisines = final_df['Cuisines'].dropna()

# Split the cuisines and flatten the list using numpy
split_cuisines = cuisines.str.split(',').explode().str.strip()

# Count the occurrences of each cuisine
top_10_cuisines = split_cuisines.value_counts().head(10)

# Convert to DataFrame for cleaner display (optional)
top_10_cuisines_df = top_10_cuisines.reset_index()
top_10_cuisines_df.columns = ['Cuisine', 'Count']

# Display the result
print(top_10_cuisines_df)
```
