```python
import seaborn as sns
import pandas as pd
```

```python
# sales=sns.load_dataset('tips')
sales=pd.read_excel('sales.xlsx')
sns.set_theme(style="whitegrid", palette="dark")
```

```python
sales.head()
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, hue='Order Priority');
# sns.set_style('darkgrid')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, hue='Order Priority', style='Ship Mode')
```

```python
sns.relplot(x='Sales', y='Profit', data=sales, size='Discount', sizes=(25,200))
```

```python
sns.relplot(x='Order Date', y='Sales' , data=sales.head(20), kind='line')
```

```python
sales['year']=sales['Order Date'].dt.year
```

```python
sales.year.unique()
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line')
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', errorbar='sd');
#ci stands for confidence interval
#since ci is deprecated we will use errorbar=None for same effect
#set errorbar='sd' for standard deviation
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', estimator=sum);
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode');
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', style="Product Category", errorbar=None);
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', col='Product Category', errorbar=None);
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', col='Region', errorbar=None);
```

```python
sns.relplot(x='year', y='Sales', data=sales, kind='line', hue='Ship Mode', col='Region', errorbar=None, col_wrap=2);
```
