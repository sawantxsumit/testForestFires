# Hyperparameter tuning and cross validation


Hyperparameter tuning is the process of choosing the best set of hyperparameters for a machine learning model to improve its performance.

1. C  : Inverse of regularization strength. Smaller C means stronger regularization (simpler model).
2. penalty :	Type of regularization. Common options: 'l1', 'l2', 'elasticnet', or None.
3. solver  :	Algorithm used for optimization. Options: 'liblinear', 'saga', 'lbfgs', etc. Some solvers support only certain penalties.
4. max_iter  :	Maximum number of iterations taken for convergence. Increase if the model doesn’t converge.

```python
model=LogisticRegression()
penalty=['l1' , 'l2','elasticnet']
c_value=[100,10,1.0,0.1 , 0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag' , 'saga']
mIter=[100,200]
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

Regularization (C, penalty) controls overfitting.

Solver affects convergence and speed.

Right combination gives better accuracy, lower loss, and generalizes well to unseen data.

```python
params=dict(penalty=penalty,C=c_value , solver=solver, max_iter=mIter)
```

```python
# Grid search CV
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model , param_grid=params , scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(x_test)
```

```python
score=accuracy_score(y_test, y_pred)
print(score)
cm=confusion_matrix(y_test, y_pred)
print(cm)
report=classification_report(y_test, y_pred)
print(report)
```

Grid Search

Try all combinations of hyperparameters from a predefined grid.

Very accurate but computationally expensive.

Example: Try every combination of max_depth = [3, 5, 7] and min_samples_split = [2, 4].

# Randomized search CV

```python

model=LogisticRegression()

from sklearn.model_selection import RandomizedSearchCV
rcv=RandomizedSearchCV(estimator=model ,param_distributions=params, n_jobs=-1 )
```

```python
rcv.fit(x_train, y_train)
```

```python
rcv.best_score_
```

```python
rcv.best_params_
```

```python
y_pred=rcv.predict(x_test)
```

```python
score=accuracy_score(y_test, y_pred)
print(score)
cm=confusion_matrix(y_test, y_pred)
print(cm)
report=classification_report(y_test, y_pred)
print(report)
```

 # logistic regression for multi class classification

```python
# create the database
x, y= make_classification(n_samples=1000 , n_features=10 , n_informative=3,n_classes=3, random_state=22)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test= train_test_split(x,y,test_size=0.3 , random_state=33)
```

```python
# logreg=LogisticRegression(multi_class='ovr')
# the above method is depricated
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression

logreg = OneVsRestClassifier(LogisticRegression())
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)

```

```python
y_pred
```

```python
score=accuracy_score(y_test, y_pred)
print(score)
cm=confusion_matrix(y_test, y_pred)
print(cm)
report=classification_report(y_test, y_pred)
print(report)
```

```python
# hyperparamter tuning
logreg = OneVsRestClassifier(LogisticRegression(solver='saga', max_iter=10000))
```

from scipy.stats import loguniform
-> This imports a logarithmic uniform distribution from scipy, which is used to sample values for the hyperparameter C during hyperparameter tuning.

🔍 What is loguniform(1e-4, 1e2)?
This creates a distribution of values that are:

logarithmically spaced between 1e-4 (0.0001) and 1e2 (100)

So instead of sampling linearly (like 1, 2, 3...), it samples values like:
0.0001, 0.001, 0.01, 0.1, 1, 10, 100


📘 Why loguniform for C?
C is the inverse of regularization strength in Logistic Regression:

High C → low regularization (model fits more to the training data)

Low C → high regularization (model is penalized more for complexity)

👉 The range of good C values varies exponentially, so we use a log scale to explore the space efficiently.

```python
from scipy.stats import loguniform

params = {
    'estimator__C': loguniform(1e-4, 1e2),  
    'estimator__penalty': ['l2', 'l1'],                 
}
```

```python
random_search = RandomizedSearchCV(
    estimator=logreg,
    param_distributions=params,
    n_iter=20,
    scoring='accuracy',
    cv=5,
    random_state=42,
    n_jobs=-1
)
```

```python
random_search.fit(x_train, y_train)
```

```python
random_search.best_params_
```

```python
random_search.best_score_
```

```python
# Predict with best estimator
best_model = random_search.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
