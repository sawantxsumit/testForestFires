```python
x=df.iloc[:, :-1] #dependent features
y=df.iloc[ :, -1] # independent features
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test= train_test_split(x,y, test_size=0.35 , random_state=22)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()
```

```python
x_train=scalar.fit_transform(x_train)
```

```python
x_test=scalar.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Model training

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# prediction for test data
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
```

```python
print(mean_squared_error(y_test , y_pred))
print(mean_absolute_error(y_test , y_pred))
print(np.sqrt(mean_squared_error(y_test , y_pred)))
```

```python
# rsqaure and adjusted r squared
from sklearn.metrics import r2_score
score= r2_score(y_test , y_pred)
print(score)
```

```python
# display the adjusted r squared
ar2= 1-(1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
ar2
```
