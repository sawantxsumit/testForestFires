```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
x= 6* np.random.rand(100 , 1) -3
y=0.5 * x**2 +1.5*x +2 +np.random.rand(100, 1)
# quadratic equation used y=0.5x^2+ 1.5x+2+outliers
plt.plot(x,y, 'g')
plt.xlabel('x dataset')
plt.ylabel('y dataset')
```

```python
plt.scatter(x,y, color='g')
plt.xlabel('x dataset')
plt.ylabel('y dataset')
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test= train_test_split(x,y, test_size=0.2 , random_state=22)
```

```python
# implement a simple linear regression 
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
x_train
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test , lr.predict(x_test))
print(score)
#acuracy is low 
```

```python
# visualization
plt.plot(x_train , lr.predict(x_train), color='r')
plt.scatter(x_train , y_train)
plt.xlabel('X')
plt.ylabel('Y')

         
```

```python
# now solving this using polynomial regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly= PolynomialFeatures(degree=2 , include_bias=True)
#inlcude bias means that 1 will also be included
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)
```

```python
x_train_poly
# output format is [bias , x1 feature , x1^2]
```

```python
from sklearn.metrics import r2_score
lr.fit(x_train_poly , y_train)
y_pred=lr.predict(x_test_poly)
score=r2_score(y_test ,y_pred)
print(score)
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
plt.scatter(x_train , lr.predict(x_train_poly))
plt.scatter(x_train ,y_train)
```

```python
poly= PolynomialFeatures(degree=3 , include_bias=True)
#inlcude bias means that 1 will also be included
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)
```

```python
x_train_poly
```

```python
from sklearn.metrics import r2_score
lr.fit(x_train_poly , y_train)
y_pred=lr.predict(x_test_poly)
score=r2_score(y_test ,y_pred)
print(score)
```

```python
# prediction on new data
x_new= np.linspace(-3,3,200).reshape(200,1)
x_new_poly= poly.transform(x_new)
x_new_poly
```

```python
y_new = lr.predict(x_new_poly)
plt.plot(x_new , y_new , 'r-' , linewidth=2  ,label='Predictions')
plt.plot(x_train , y_train , 'b.' , label='Training  points')
plt.plot(x_test , y_test , 'g.' , label='Testing  points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Pipeline comcept

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    x_new=np.linspace(-3,3,200).reshape(200,1)
    x_new_poly= poly.transform(x_new)
    poly_features=PolynomialFeatures(degree=degree , include_bias=True)
    lr=LinearRegression()
    poly_regression=Pipeline([
        ('poly_features' , poly_features), 
        ('lin_regression' , lr)
    ])
    poly_regression.fit(x_train , y_train)
    y_pred_new=poly_regression.predict(x_new)
    #plotting prediction line
    plt.plot(x_new , y_pred_new , 'r' , label='Degree'+ str(degree) , linewidth=2 )
    plt.plot(x_train , y_train  , 'b.' , linewidth=3)
    plt.plot(x_test, y_test , 'g.' , linewidth=3)
    plt.legend(loc='upper left')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.axis([-4 , 4 ,0, 10])
    plt.show()
```

```python
poly_regression(2)
```

```python
poly_regression(10)
#starts overfitting for high degree values
```

```python
poly_regression(6)
```
