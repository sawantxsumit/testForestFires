```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test= train_test_split(x,y , test_size=0.3,random_state=33)
```

```python
x_train.head()
```

```python
# feature encoding
# 1. label encoding 2. onehot encoding
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
le1.fit_transform(x_train['sex'])
le2.fit_transform(x_train['smoker'])
le3.fit_transform(x_train['time'])
```

```python
x_train['sex']=le1.fit_transform(x_train['sex'])
x_train['smoker']=le2.fit_transform(x_train['smoker'])
x_train['time']=le3.fit_transform(x_train['time'])
```

```python
x_train.head()
```

```python
x_test['sex']=le1.transform(x_test['sex'])
x_test['smoker']=le2.transform(x_test['smoker'])
x_test['time']=le3.transform(x_test['time'])
```

```python
x_test.head()
```

```python
#one hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('onehot' , OneHotEncoder(drop='first'), [3])] , remainder='passthrough')
```

```python
ct.fit_transform(x_train)
```

```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
x_train=ct.fit_transform(x_train)
x_test=ct.fit_transform(x_test)
```

```python
# support vector regression
from sklearn.svm import SVR
```

```python
svr=SVR()
```

```python
svr.fit(x_train,y_train)
y_pred= svr.predict(x_test)
```

```python
from sklearn.metrics import r2_score , mean_absolute_error
print(r2_score(y_test , y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
# hyper parameter tuning using gridsearch
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1 , 1, 10 ,100 , 1000] ,
           'gamma' : [1,0.1 , 0.01 , 0.001 , 0.0001],
           'kernel' :['rbf']}
```

```python
grid=GridSearchCV(SVR(), param_grid=param_grid ,refit=True , cv=5 , verbose=3 )
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid_predictions=grid.predict(x_test)
```

```python
print(r2_score(y_test , grid_predictions))
print(mean_absolute_error(y_test, grid_predictions))
```
