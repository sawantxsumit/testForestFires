```python
car_missing2=pd.read_csv('data/car-sales-extended-missing-data.csv')
```

```python
car_missing2.isna().sum()
```

```python
car_missing2.dropna(subset=['Price'] , inplace=True)
car_missing2.isna().sum()
```

```python
#split into x and y
x=car_missing2.drop('Price' , axis=1)
y=car_missing2['Price']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

#fill categorical values with 'missing' and numerical values with mean
cat_imputer=SimpleImputer(strategy='constant' , fill_value='missing')
door_imputer=SimpleImputer(strategy='constant' , fill_value=4)
num_imputer=SimpleImputer(strategy='mean')

#define columns
cat_features=['Make' , 'Colour']
door_features=['Doors']
num_features=['Odometer (KM)']

#create an imputer 
imputer= ColumnTransformer([
    ('cat_imputer', cat_imputer, cat_features),
    ('door_imputer' , door_imputer, door_features),
    ('num_imputer' , num_imputer ,num_features)
])

#transform the data
filled_x=imputer.fit_transform(x)
filled_x
```

```python
car_sales_filled=pd.DataFrame(filled_x , columns=['Make' , 'Colour', 'Doors','Odometer'])
car_sales_filled
```

```python
car_sales_filled.isna().sum()
```

```python
#label encoding categorical features
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

category_features=['Make' , 'Colour', 'Doors']
oneHot=OneHotEncoder()
optimus=ColumnTransformer([('One_hot', oneHot, category_features)], remainder='passthrough')
optimus_c=optimus.fit_transform(car_sales_filled)
optimus_c
```

```python
#now let's fit a model
np.random.seed(22)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test= train_test_split(optimus_c , y , test_size=0.2)
model= RandomForestRegressor()
model.fit(x_train , y_train)
model.score(x_test , y_test)
```

```python
# 2.1 picking machine learning model for a regression problem
# import california housing dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
#target variable is the median house value for California districts,expressed in hundreds of thousands of dollars ($100,000).

housing_df=pd.DataFrame(housing['data'] , columns=housing['feature_names'])
housing_df['target']=pd.Series(housing['target'])
housing_df.head()
```

```python
# number of samples
len(housing_df)
```

Concept	      :        Simple Meaning

Regression	  :    Predicting a number (like test marks)

Classification	:  Predicting a category (like dog vs cat)

Ridge	   :       A method to avoid overfitting by adding a penalty

Alpha (λ)    :     Controls how strong the penalty is

Why it's useful?  :	Makes models simpler, better, and less “jumpy”


Ridge Regression is a type of linear regression that includes L2 regularization.
It is used to prevent overfitting in cases where there are:
 * Many features (especially if they are correlated)

 * Small or noisy datasets

It works by penalizing the magnitude of the coefficients (weights) in the model, discouraging large values.


                                                                               L2 regularization is a technique used in machine learning to prevent overfitting by discouraging large values in model coefficients (weights).

📌 Why Do We Need It?
When a model is too flexible (especially with many features), it can memorize the training data — this is called overfitting.
It performs well on training data but poorly on new (test) data.
Regularization helps by adding a penalty to the model's loss function, making it prefer simpler models with smaller weights.

```python
from sklearn.linear_model import Ridge
np.random.seed(44)
#create the data
x=housing_df.drop('target' , axis=1)
y=housing_df['target']

#split the data
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size=0.2)

#instantiate  ridge model
model=Ridge()
model.fit(x_train , y_train)

model.score(x_test , y_test)
```

```python
# How to improve the model
# what if ridge was'nt working
from sklearn.ensemble import RandomForestRegressor
np.random.seed(99)
x=housing_df.drop('target' , axis=1)
y=housing_df['target']

#split the data
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size=0.2)

rmodel=RandomForestRegressor(n_estimators=100)
rmodel.fit(x_train , y_train)
rmodel.score(x_test , y_test)
```
